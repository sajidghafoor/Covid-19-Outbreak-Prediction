```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Total cases per day
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=pak_case)
plt.show()
```

```python
#Making a dataframe for last 5 days
pak_last_5_days=pak_case.tail()
```

```python
#Total cases in last 5 days
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=pak_last_5_days)
plt.show()
```

```python
#Total tests per day
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_tests",data=pak_case)
plt.show()
```

```python
#Total tests in last 5 days
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_tests",data=pak_last_5_days)
plt.show()
```

```python
#Understanding cases of India, China and Pakistan
pk_chn_ind=covid[(covid["location"] =="Pakistan") | (covid["location"] =="China") | (covid["location"]=="India")]
```

```python
#Plotting growth of cases across China, India and Japan
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="location",y="total_cases",data=pk_chn_ind)
plt.show()
```

```python
#Getting latest data
last_day_cases=covid[covid["date"]=="2020-05-24"]
last_day_cases
```

```python
#Sorting data w.r.t total_cases
max_cases_country=last_day_cases.sort_values(by="total_cases",ascending=False)
max_cases_country
```

```python
#Top 5 countries with maximum cases
max_cases_country[1:6]
```

```python
#Making bar-plot for countries with top cases
sns.barplot(x="location",y="total_cases",data=max_cases_country[1:6],hue="location")
plt.show()
```
